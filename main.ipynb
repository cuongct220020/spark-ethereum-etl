```python
from pyspark.sql import SparkSession

# Khởi tạo SparkSession
spark = SparkSession.builder \
   .appName("EthereumAnalysis") \
   .master("spark://spark-master:7077") \
   .config("spark.eventLog.enabled", "true") \
   .config("spark.eventLog.dir", "file:///opt/spark/spark-events") \
   .getOrCreate()
```

```python
from pyspark.sql.types import StructType, StructField, StringType, LongType, DecimalType

# Schema cho transactions.csv
tx_schema = StructType()

# Schema cho token_transfers.csv
transfer_schema = StructType()
```

```python
# Đọc Transactions
df_tx = spark.read \
   .format("csv") \
   .option("header", "true") \
   .schema(tx_schema) \
   .load("/opt/spark/data/transactions.csv")

# Đọc Token Transfers
df_transfers = spark.read \
   .format("csv") \
   .option("header", "true") \
   .schema(transfer_schema) \
   .load("/opt/spark/data/token_transfers.csv")
```
